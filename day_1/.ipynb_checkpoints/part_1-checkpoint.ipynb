# What is the problem?
The problem provides a calibration document where each line contains a sequence of characters
I have to combine the first digit and last digit of each line to form a two digit number.
Then I have to add up all these values (2 digits) and get the total sum to find the solution to this prblem.

# advent of code examples:
* 1abc2: first digit = 1, last digit = 2 -> number of two digits = 12
* pqr3stu8vwx: first digit = 3, last digit = 8 -> number of two digits = 38
* a1b2c3d4e5f: first digit = 1, last digit = 5 -> number of two digits = 15
* treb7uchet: first digit = 7, last digit = 7 -> number of two digits = 77 

# What to do?
* solve the problem with the example of advent of code, to see if the final result is 142, it's a way to understand if I'm doing the right steps;
* read the calibration document line by line -> search how to read a text file in python;
* for each line, extract the first and last digit;
* combine the two digits to form a two-digit number;
* add up all these two digit values;
* see total sum.

```python
#Create a file.txt when I can put the inputs of the problem.
read_file = open("input_part_1.txt", "r")
content = read_file.read()
print(content)
read_file.close()

#Pass the text file to an array: transform "content" variable into an array of strings.
array_of_strings = content.split()

#Now I have an array of strings, beacuse of the split().
#I want to read each string and remove the characters that are letters to leave only the characters that are numbers.
array_of_ch_numbers = []

for string in array_of_strings:
    numbers = ''.join(ch for ch in string if ch.isdigit())
    array_of_ch_numbers.append(numbers)
    

#Now I have an array of string that contains only number characters.
#What I want now is to read each string and only have 2 characters (the first and last), for example if the string = "a1b2c3d4e5f" the output of that string will be ["15"].
#So I have to create a new list to store character pairs.

array_of_two_ch_numbers = []

for string in array_of_ch_numbers:
    first_ch = string[0]
    last_ch = string[-1]

    two_ch_numbers = first_ch + last_ch
    

    array_of_two_ch_numbers.append(two_ch_numbers)


#Transform array of strings into an array of numbers.
array_of_numbers = []

for two_digit in array_of_two_ch_numbers:
    array_of_numbers.append(int(two_digit))

#Calculate the sum os numbers inside array_of_numbers:
sum_of_array_of_numbers = sum(array_of_numbers)
print(sum_of_array_of_numbers)
```
